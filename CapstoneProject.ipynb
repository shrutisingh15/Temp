```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
plt.style.use('ggplot')
# library to determine languages used in tweet messages
import langid
```

```python
# Reading the data in Pandas DataFrame
data= pd.read_csv('tweets.csv')
```

```python
# Lets have a glimpse of the data
data.head()
```

```python
# Using list comprehension to classify the language of the tweets
langs = [langid.classify(tweet) for tweet in data['tweets']]
```

```python
# Assigning it to a pandas dataframe
langs_data = pd.DataFrame(langs, columns = ['language','count'])
```

```python
langs_data.head()
```

```python
## Top 10 most commonly used languages
plot_data = langs_data['language'].value_counts().head(10)
```

```python
colors = sns.color_palette('Set1',10)
plot_data.plot(kind = "bar", figsize = (12,9), color = colors, fontsize = 12, rot = 45, title = "Top 10 most common languages")
```

The above graph shows that english is the most common language used in this data, the second most common being arabic. So, what are they talking in english. Lets take a look at the word cloud of the tweets...

```python
import re
from sklearn.feature_extraction.text import TfidfVectorizer
from wordcloud import WordCloud
```

```python
extrawords = re.compile("al|RT|\n|&.*?;|http[s](?://(?:[a-zA-Z]|[0-9]|[$-_@.&+]|[!*\(\),]|(?:%[0-9a-fA-F][0-9a-fA-F]))+)*")
tweet_text = [extrawords.sub(" ", t) for t in data['tweets']]
```

```python
vectorizer = TfidfVectorizer(stop_words = 'english', ngram_range=(1,2), max_df=0.5)
tfidf = vectorizer.fit_transform(tweet_text)
```

```python
terms = vectorizer.get_feature_names()
```

```python
wc = WordCloud(height = 1200, width = 1200, max_words = 1200).generate(" ".join(terms))
plt.figure(figsize = (10,10))
plt.imshow(wc)
plt.axis("off")
plt.show()
```

```python
from sklearn.decomposition import NMF
nmf = NMF(n_components = 10).fit(tfidf)
for idx, topic in enumerate(nmf.components_):
    print "Topic #%d:" %idx
    print " ".join([terms[i] for i in topic.argsort()[:-10 -1: -1 ]])
```
